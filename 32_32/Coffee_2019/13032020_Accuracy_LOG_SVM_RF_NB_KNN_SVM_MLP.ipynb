```python
#print("aaa")

COMMON_SIZE = (100, 100)
def image_to_matrix(filename):
	"""
	takes a filename and turns it into a numpy array of RGB pixels
	"""
	img = Image.open(filename)
	img = img.resize(COMMON_SIZE)
	img = np.asarray(img)
	return img

def flatten_matrix(matrix):
	"""
	takes in an (m, n) numpy array and flattens it
	into an array of shape (1, m * n * 3)
	"""
	s = matrix.shape[0] * matrix.shape[1] * 3
	mat = matrix.reshape(1,s)
	#print(mat.shape)
	return mat[0]
#60:30:10
#01
classes = ['Good', 'Peaberry']
#32_32 image pixel
#train_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/32_32/Coffee_2019/Train/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/32_32/Coffee_2019/Train/Peaberry/"]
#64_64 image pixel
#train_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/64_64/Coffee_2019/Train/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/64_64/Coffee_2019/Train/Peaberry/"]
#128_128 image pixel
#train_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/128_128/Coffee_2019/Train/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/128_128/Coffee_2019/Train/Peaberry/"]
#256_256 image pixel
train_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/256_256/Coffee_2019/Train/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/256_256/Coffee_2019/Train/Peaberry/"]

train_images = []
train_labels = []
# Retrieval of path of all pictures from "train" folder
for directory in train_image_dirs:
	temp_list = [directory+ f for f in os.listdir(directory)]
	train_labels.extend(len(temp_list)*[classes[train_image_dirs.index(directory)]])
	train_images.extend(temp_list)

# Creation of the whole training dataset
train_data = []
for image in train_images:
	image = image_to_matrix(image)
	image = flatten_matrix(image)
	train_data.append(image)
train_data = np.array(train_data)

# Creation of the whole test set
test_labels = []

#60:30:10
#32_32 image pixel
#test_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/32_32/Coffee_2019/Test/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/32_32/Coffee_2019/Test/Peaberry/"]
#64_64 image pixel
#test_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/64_64/Coffee_2019/Test/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/64_64/Coffee_2019/Test/Peaberry/"]
#128_128 image pixel
#test_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/128_128/Coffee_2019/Test/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/128_128/Coffee_2019/Test/Peaberry/"]
#256_256 image pixel
test_image_dirs = ["C:/Users/Hira/Desktop/Coffee_2019_13032020/256_256/Coffee_2019/Test/Good/", "C:/Users/Hira/Desktop/Coffee_2019_13032020/256_256/Coffee_2019/Test/Peaberry/"]

test_images = []
for directory in test_image_dirs:
	temp_list = [directory+ f for f in os.listdir(directory)]
	test_labels.extend(len(temp_list)*[classes[test_image_dirs.index(directory)]])
	test_images.extend(temp_list)

test_data = []
for image in test_images:
	image = image_to_matrix(image)
	image = flatten_matrix(image)
	test_data.append(image)
test_data = np.array(test_data)
```

```python
#Support Vector Machine (SVM) 
```

```python
#Linear SVM
clf = svm.SVC(kernel = 'linear',gamma = 50, C=1)

#non-linear SVM
#clf = svm.NuSVC(nu=0.4, kernel="rbf", gamma=0.1)
#clf = svm.NuSVC(gamma='auto')
clf.fit(train_data, train_labels)
print("SVM average score: %f" % clf.score(test_data, test_labels))

predicted_label = clf.predict(test_data)

Accuracy_Score = accuracy_score(test_labels, predicted_label)
Precision_Score = precision_score(test_labels, predicted_label,  average="macro")
Recall_Score = recall_score(test_labels, predicted_label,  average="macro")
F1_Score = f1_score(test_labels, predicted_label,  average="macro")

print('Average Accuracy: %0.2f +/- (%0.1f) %%' % (Accuracy_Score.mean()*100, Accuracy_Score.std()*100))
print('Average Precision: %0.2f +/- (%0.1f) %%' % (Precision_Score.mean()*100, Precision_Score.std()*100))
print('Average Recall: %0.2f +/- (%0.1f) %%' % (Recall_Score.mean()*100, Recall_Score.std()*100))
print('Average F1-Score: %0.2f +/- (%0.1f) %%' % (F1_Score.mean()*100, F1_Score.std()*100))

print("")
#https://en.wikipedia.org/wiki/Confusion_matrix
CM = confusion_matrix(test_labels, predicted_label)

print ('Confusion Matrix :')
print(CM)
for i in predicted_label:
	print(i)	   
```

```python
def plot_decision_function(model):
    _x0 = np.linspace(-5.0, 5.0, 100)
    _x1 = np.linspace(-5.0, 5.0, 100)
    x0, x1 = np.meshgrid(_x0, _x1)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = model.predict(X).reshape(x0.shape)
    y_decision = model.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision,levels=10,alpha=0.3)
 
def plot_dataset(X,y):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bo", ms=10)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "r^", ms=10)
    plt.xlabel("A")
    plt.ylabel("B")
```
